```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("carprices.csv")
data
```

```python
dummies=pd.get_dummies(data.carModel)
dummies
```

```python
merge=pd.concat([data,dummies],axis="columns")
merge
```

```python
new=merge.drop(["carModel","Mercedez Benz C class"],axis="columns")
new
```

```python
x=new.drop(["SellPrice"],axis="columns")
x
```

```python
y=new.iloc[:,:2]
Y=y.drop(["Mileage"],axis="columns")
Y
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x,Y)
```

```python
pred=reg.predict([[22500,1,0,1]])
pred
```

```python
reg.score(x,Y)
```
